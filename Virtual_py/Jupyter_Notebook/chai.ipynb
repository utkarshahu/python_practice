# 🐍 Python Virtual Environment Setup Guide

This guide helps you create a **virtual environment in Python** to isolate your project dependencies. It works on both **Windows** and **macOS/Linux**.

---

## 🔧 Prerequisites

- Python (version 3.5 or above) should be installed.
- Check if Python is installed:

```bash
python --version
# or
python3 --version
````

---

## 📁 Step 1: Open Terminal or Command Prompt

* **Windows**: Use `cmd`, PowerShell, or the terminal in VS Code.
* **macOS/Linux**: Open your system terminal.

---

## 📂 Step 2: Navigate to Your Project Directory

```bash
cd path/to/your/project-folder
```

Example:

```bash
cd D:\projects\myapp
```

---

## 🔄 Step 3: Create a Virtual Environment

### ✅ Windows

```bash
python -m venv env
```

### ✅ macOS/Linux

```bash
python3 -m venv env
```

> A folder named `env` will be created in your project directory.

---

## 🚀 Step 4: Activate the Virtual Environment

### ✅ Windows

```bash
env\Scripts\activate
```

### ✅ macOS/Linux

```bash
source env/bin/activate
```

> After activation, your terminal prompt will change to:
> `(env) C:\Users\Utkarsh\myapp>`
> This means the environment is active.

---

## 📦 Step 5: Install Project Dependencies

Install packages inside the virtual environment using `pip`:

```bash
pip install flask
```

---

## ❌ Step 6: Deactivate the Virtual Environment

To deactivate the environment and return to the global Python:

```bash
deactivate
```

---

## ✅ Notes

* Use this setup to avoid dependency conflicts across projects.
* You can add `env/` to `.gitignore` to avoid uploading it to version control systems like GitHub.

---

Happy Coding! 🚀
