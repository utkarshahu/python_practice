---

## 🧠 What is Conda?

### 🔹 Conda is:

* A **package manager** (like `pip`)
* A **virtual environment manager** (like `venv`)
* It works with **Python**, **R**, and other languages.
* Comes from the **Anaconda** and **Miniconda** distributions.

> 📦 Think of Conda as a tool that installs **Python packages + their dependencies** in **clean, isolated environments**.

---

## ✅ Why Use Conda?

* It handles **both Python packages** and **non-Python dependencies** (like C libraries).
* Ideal for **Data Science**, **Machine Learning**, and **AI** projects.
* Makes it easy to switch between different versions of Python and packages.

---

## 🧰 How to Install Conda

You have **2 options**:

### Option 1: Install **Anaconda** (Big + Includes tools like Jupyter, Pandas, etc.)

* Size: \~500 MB+
* Best for beginners who want **everything included**

🔗 Download from: [https://www.anaconda.com/download](https://www.anaconda.com/download)

### Option 2: Install **Miniconda** (Lightweight)

* Size: \~50 MB
* Just Conda + Python
* Best if you want to install packages manually

🔗 Download from: [https://docs.conda.io/en/latest/miniconda.html](https://docs.conda.io/en/latest/miniconda.html)

---

## 💻 How to Install (Step-by-Step)

### ✅ Step 1: Download the Installer

Choose the correct file for your system:

* **Windows:** `Miniconda3 Windows 64-bit .exe`
* **macOS/Linux:** `.sh` installer

### ✅ Step 2: Run the Installer

#### On Windows:

* Double-click the `.exe` file
* Click **Next**, accept the license, and install

#### On macOS/Linux:

Run in terminal:

```bash
bash Miniconda3-latest-Linux-x86_64.sh
```

Follow the prompts.

### ✅ Step 3: Check if Conda is Installed

```bash
conda --version
```

You should see something like:

```
conda 24.1.2
```

---

## ⚙️ How to Use Conda

### 🌀 Step 1: Create a Virtual Environment

```bash
conda create --name myenv python=3.11
```

* `myenv` = name of your environment
* `python=3.11` = optional, you can set any version

### 🚀 Step 2: Activate the Environment

```bash
conda activate myenv
```

Now your terminal will look like:

```bash
(myenv) C:\Users\Utkarsh>
```

### 📦 Step 3: Install Packages

```bash
conda install numpy
```

You can also install from specific channels like `conda-forge`:

```bash
conda install -c conda-forge matplotlib
```

### 📋 Step 4: List All Environments

```bash
conda env list
```

### ❌ Step 5: Deactivate the Environment

```bash
conda deactivate
```

---

## 🗑️ How to Remove an Environment

```bash
conda remove --name myenv --all
```

---

## 🔁 Bonus: Convert Environment to `requirements.txt`

```bash
conda list --export > requirements.txt
```

Or better, use:

```bash
conda env export > environment.yml
```

Then recreate with:

```bash
conda env create -f environment.yml
```

## ❓ Pip vs Conda - Quick Comparison

| Feature          | pip            | conda                 |
| ---------------- | -------------- | --------------------- |
| Python packages  | ✅ Yes          | ✅ Yes                 |
| Non-Python libs  | ❌ No           | ✅ Yes                 |
| Virtual env mgmt | Needs `venv`   | Built-in              |
| Speed            | Sometimes slow | Faster for large pkgs |
