
## 🧠 What is Jupyter Notebook?

> **Jupyter Notebook** is an interactive web-based environment to write, run, and visualize Python code.

It allows you to:

* ✍️ Write Python code in blocks called **cells**
* 📊 Show charts, tables, and data
* 📚 Add headings, notes, and explanations using **Markdown**
* 🧪 Run code step-by-step for experimentation

> It’s the **perfect tool for Data Science, Machine Learning, and learning Python interactively**.

---

## ✅ Features of Jupyter Notebook

| Feature           | Description                             |
| ----------------- | --------------------------------------- |
| 🔢 Code Cells     | Write & execute Python code             |
| ✍️ Markdown Cells | Add notes, headings, equations          |
| 📊 Visual Output  | Show plots, tables, graphs              |
| 💾 Save Work      | Saves as `.ipynb` files                 |
| 🔄 Export Options | Convert to `.py`, `.pdf`, `.html`, etc. |

---

## 🧰 How to Install Jupyter Notebook

You can install it in 2 main ways:

### 🔸 Option 1: Using pip (recommended if you already have Python)

```bash
pip install notebook
```

### 🔸 Option 2: Using Conda (if you have Anaconda or Miniconda)

```bash
conda install -c conda-forge notebook
```

---

## 🚀 How to Start Jupyter Notebook

### Step 1: Open Terminal or Command Prompt

Go to your project folder:

```bash
cd path/to/your/folder
```

### Step 2: Start the Notebook Server

```bash
jupyter notebook
```

### Step 3: It will open in your browser

URL will look like:

```
http://localhost:8888/tree
```

Here you can:

* Click **New > Python 3 Notebook**
* Start writing and running code

## 🖼️ Example: Data Visualization

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Line Chart")
plt.show()
```

🟢 Output will be shown **below the cell** — no terminal or print needed.

## 🧪 Shortcuts You’ll Love

| Shortcut        | Action                       |
| --------------- | ---------------------------- |
| `Shift + Enter` | Run cell                     |
| `A`             | Insert cell **above**        |
| `B`             | Insert cell **below**        |
| `M`             | Convert cell to **Markdown** |
| `Y`             | Convert cell to **Code**     |
| `Ctrl + S`      | Save notebook                |
